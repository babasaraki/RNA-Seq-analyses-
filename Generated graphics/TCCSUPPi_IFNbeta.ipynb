```python
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("TCCSUPPi_IFNbeta.csv")
df
```

```python
sns.barplot(x='IFNβ (pg/mL)', y='TCCSUP control vs TCCSUPPi cells', hue='TCCSUP_cells', data=df)
```

```python
sns.barplot(x='IFNβ (pg/mL)', y='TCCSUP control vs TCCSUPPi cells', hue='TCCSUP_cells', capsize=0.05, data=df)
```

```python
sns.barplot(x='IFNβ (pg/mL)', y='TCCSUP control vs TCCSUPPi cells', hue='TCCSUP_cells',
            palette="mako", capsize=0.05, data=df)
```

```python
sns.barplot(x='IFNβ (pg/mL)', y='TCCSUP control vs TCCSUPPi cells', hue='TCCSUP_cells',
            palette="mako", capsize=0.05, data=df)
plt.savefig('TCCSUPPi_IFNbeta.png', dpi=300)
```

```python
sns.barplot(x='IFNβ (pg/mL)', y='TCCSUP control vs TCCSUPPi cells', hue='TCCSUP_cells',
            palette="mako", capsize=0.05, data=df)
plt.savefig('TCCSUPPi_IFNbeta.pdf', bbox_inches='tight', pad_inches=0.1, dpi=300)
```
