```python
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("qPCR_TCCSUP.csv")
df
```

```python
sns.barplot(x='Fold change (log2R)', y='Genes', hue='Methods', data=df)
```

```python
sns.barplot(x='Fold change (log2R)', y='Genes', hue='Methods', capsize=0.1, data=df)
```

```python
sns.barplot(x='Fold change (log2R)', y='Genes', hue='Methods',
            palette="terrain", capsize=0.1, data=df)
plt.savefig('qPCR_TCCSUP.png', dpi=300)
```

```python
sns.barplot(x='Fold change (log2R)', y='Genes', hue='Methods',
            palette="terrain", capsize=0.1, data=df)
plt.savefig('qPCR_TCCSUP.pdf', dpi=300)
```
