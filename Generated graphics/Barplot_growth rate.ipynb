```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Growth rate.csv")
df
```

```python
sns.barplot(x='Cell_lines', y='Number of Days', data=df)
```

```python
sns.catplot(x='Cell_lines', y='Number of Days', kind="box", data=df)
```

```python
sns.catplot(x='Cell_lines', y='Number of Days', kind="violin", data=df)
```

```python
sns.barplot(x='Cell_lines', y='Number of Days', capsize=0.1, data=df)
plt.savefig('Growth_rate.png', dpi=300)
plt.savefig('Growth_rate.pdf', dpi=300)
```
